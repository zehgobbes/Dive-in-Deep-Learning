# 4.6.2. Reutilização de conjunto de testes

De certa forma, agora você está preparado para ter sucesso na condução de pesquisas empíricas de aprendizado de máquina. Quase todos os modelos práticos são desenvolvidos e validados com base no desempenho do conjunto de testes e agora você é um mestre no conjunto de testes.

Depois que as informações do conjunto de testes vazarem para o modelador, __elas nunca mais poderão ser um conjunto de testes verdadeiro no sentido mais estrito.__ Este problema é denominado overfitting adaptativo e emergiu recentemente como um tópico de intenso interesse para teóricos da aprendizagem e estatísticos ( Dwork et al. , 2015 ). Felizmente, embora seja possível vazar todas as informações de um conjunto de resistência e os piores cenários teóricos sejam sombrios, essas análises podem ser muito conservadoras. 
 
__Na prática, tome cuidado para criar conjuntos de testes reais, para consultá-los com a menor frequência possível, para levar em conta múltiplos testes de hipóteses ao relatar intervalos de confiança e para aumentar sua vigilância de forma mais agressiva quando os riscos são altos e o tamanho do seu conjunto de dados é pequeno. Ao executar uma série de desafios de benchmark, muitas vezes é uma boa prática manter vários conjuntos de testes para que, após cada rodada, o conjunto de testes antigo possa ser rebaixado a um conjunto de validação.__

# 4.6.3. Teoria da Aprendizagem Estatística

Simplificando, conjuntos de testes são tudo o que realmente temos e, ainda assim, esse fato parece estranhamente insatisfatório.

_Análise post-hoc refere-se à utilização secundária de dados previamente analisados em um estudo já completo, com o intuito de gerar novas hipóteses a partir de um recorte diferente do analisado primordialmente._

Com essas dúvidas em mente, você pode agora estar suficientemente preparado para ver o apelo da teoria da aprendizagem estatística , o subcampo matemático da aprendizagem de máquina cujos praticantes visam elucidar os princípios fundamentais que explicam por que/quando modelos treinados em dados empíricos podem/irão generalizar para dados invisíveis. Um dos principais objetivos dos pesquisadores de aprendizagem estatística tem sido limitar a lacuna de generalização, relacionando as propriedades da classe do modelo com o número de amostras no conjunto de dados.

Os teóricos da aprendizagem pretendem limitar a diferença entre o erro empírico e o verdadeiro erro.

E, de facto, qualquer classificador fixo generaliza: o seu erro num conjunto de dados (anteriormente não visto) é uma estimativa imparcial do erro populacional. Mas o que podemos dizer quando um classificador é treinado e avaliado no mesmo conjunto de dados? Poderemos algum dia ter certeza de que o erro de treinamento estará próximo do erro de teste?

Num certo sentido, a classe dos memorizadores é demasiado flexível. Nenhum resultado de convergência uniforme poderia ser válido. Por outro lado, um classificador fixo é inútil – ele generaliza perfeitamente, mas não se ajusta nem aos dados de treinamento nem aos dados de teste. A questão central da aprendizagem tem sido, portanto, historicamente enquadrada como um compromisso entre classes de modelos mais flexíveis (maior variância) que se ajustam melhor aos dados de treino, mas correm o risco de sobreajuste, versus classes de modelos mais rígidos (maior viés) que generalizam bem, mas correm o risco de subajuste. Uma questão central na teoria da aprendizagem tem sido desenvolver a análise matemática apropriada para quantificar onde um modelo se situa ao longo deste espectro e fornecer as garantias associadas.



# Resumo

A maneira mais direta de avaliar um modelo é consultar um conjunto de testes composto de dados nunca antes vistos. As avaliações do conjunto de testes fornecem uma estimativa imparcial do erro verdadeiro e convergem para o valor desejado.

Na verdade, a avaliação de conjuntos de testes é a base da pesquisa moderna em aprendizado de máquina. No entanto, os conjuntos de testes raramente são conjuntos de testes verdadeiros (usados repetidamente por vários pesquisadores). Uma vez que o mesmo conjunto de testes é usado para avaliar vários modelos, o controle de falsas descobertas pode ser difícil. 

No entanto, é uma boa prática selecionar conjuntos de testes reais (ou múltiplos) e ser o mais conservador possível quanto à frequência com que eles são usados.

Na esperança de fornecer uma solução mais satisfatória, os teóricos da aprendizagem estatística desenvolveram métodos para garantir a convergência uniforme em uma classe de modelo. Se de fato o erro empírico de cada modelo converge simultaneamente para o seu erro verdadeiro, então somos livres para escolher o modelo com melhor desempenho, minimizando o erro de treinamento, sabendo que ele também terá um desempenho semelhante nos dados de validação.

Infelizmente, embora estas medidas de complexidade tenham se tornado ferramentas amplamente úteis na teoria estatística, elas revelam-se impotentes (quando aplicadas diretamente) para explicar por que as redes neurais profundas generalizam. As redes neurais profundas geralmente têm milhões de parâmetros (ou mais) e podem facilmente atribuir rótulos aleatórios a grandes coleções de pontos. No entanto, generalizam bem em problemas práticos e, surpreendentemente, muitas vezes generalizam melhor, quando são maiores e mais profundos, apesar de incorrerem em dimensões de VC mais elevadas.
